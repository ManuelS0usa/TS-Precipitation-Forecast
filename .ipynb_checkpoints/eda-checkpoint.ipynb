Forecasting

```python
years = df['ano']
```

```python
years
```

```python
monthly_precip = df
monthly_precip.drop('ano', axis=1, inplace=True)
```

```python
monthly_precip
```

```python
rows = []
months = {
    'janeiro': '01', 
    'fevereiro': '02', 
    'março': '03', 
    'abril': '04', 
    'maio': '05', 
    'junho': '06', 
    'julho': '07', 
    'agosto': '08',
    'setembro': '09',
    'outubro': '10',
    'novembro': '11',
    'dezembro': '12'
}

for idx in years.index:    
    year = years[idx]
    for column, value in monthly_precip.iteritems():
#         print(months[column] + '-' + str(year), value[idx])    
        rows.append({
            'mes': months[column] + '-' + str(year), 
            'precipitacao': value[idx]
        }) 

mydata = pd.DataFrame(rows)            
```

```python
mydata.set_index('mes')
```

```python
mydata.plot.line(x='mes', y='precipitacao', figsize=(15,8), title='TESTE')
```
