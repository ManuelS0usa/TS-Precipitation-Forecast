```python
# sum() method sums up the rows and columns of a dataframe
# axis = 1 sums up the rows
sum_per_year = pd.DataFrame(df_year_idx.sum(axis=1), columns=["Total Precipitação p/ Ano"])
print(sum_per_year)
```

```python
idx = pd.Index(years)
sum_per_year = sum_per_year.set_index(idx)
```

```python
sum_per_year
```

```python
sum_per_year.plot.bar()
```

```python
mean_per_year = pd.DataFrame(df_year_idx.sum(axis=1), columns=["Média Precipitação p/ Ano"])
idx = pd.Index(years)
mean_per_year = mean_per_year.set_index(idx)
print(mean_per_year)
```

```python
mean_per_year.plot.bar()
```
